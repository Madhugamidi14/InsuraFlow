```python
import duckdb
import pandas as pd
from itables import show
import itables.interactive

# Enable notebook mode
itables.interactive.mode = "notebook"

# Load data
df = pd.read_json("output/transformed_insurance_data.json", lines=True)

# Register and query
con = duckdb.connect()
con.register("insurance_transformed", df)

result_df = con.execute("SELECT * FROM insurance_transformed").fetchdf()

# Display using itables
show(result_df)
```
